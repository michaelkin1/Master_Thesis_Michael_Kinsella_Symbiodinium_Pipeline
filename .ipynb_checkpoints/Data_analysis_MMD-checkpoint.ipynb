```python
#start with importing the TIFF-file and checking the metadata as well as displaying the image for downstream analysis

import tifffile

# Open the TIFF image file
with tifffile.TiffFile("C:/Users/micha/Desktop/Master_Project/tiff_file_converted.tiff") as tif:
    num_frames = len(tif.pages)
    
    frame_to_display = 290
    tiff_data = tif.asarray(key=frame_to_display)

    # Display the chosen frame
    plt.imshow(tiff_data, cmap='gray')
    plt.title(f"Frame {frame_to_display + 1}")
    plt.show()
    

    # Iterate through frames/pages
    for i, page in enumerate(tif.pages):
        # Print metadata for the current frame
        print(f"Processing Frame {i + 1} of {num_frames}")
        for tag, value in page.tags.items():
            tag_name = str(tag)
            print(f"{tag_name}: {value}")
        print("-" * 50)  # Add a separator between frames
```
