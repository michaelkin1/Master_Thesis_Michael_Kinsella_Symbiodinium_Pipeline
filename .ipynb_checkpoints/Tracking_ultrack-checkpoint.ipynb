```python
import os

from pathlib import Path

import tifffile
import napari
import numpy as np
from napari.utils.notebook_display import nbscreenshot
from tqdm import tqdm
from rich.pretty import pprint
import pandas as pd

from ultrack import track, to_tracks_layer, tracks_to_zarr
from ultrack.imgproc import normalize
from ultrack.utils import estimate_parameters_from_labels, labels_to_edges
from ultrack.utils.array import array_apply
from ultrack.config import MainConfig
import ultrack.core.export as export
```

```python
labels = tifffile.imread(r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Filtered_labels\filtered_labels_cropped_trap_1.tiff")
cropped_trap = tifffile.imread(r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Cropped_traps_directory\cropped_trap_1.tiff")
```

```python
viewer = napari.Viewer()
```

```python
viewer.add_image(cropped_trap, name="trap", colormap='gray')
viewer.add_labels(labels)
```

```python
detection, edges = labels_to_edges(labels, sigma=4.0)
```

```python
viewer.add_image(detection, visible=False)
viewer.add_image(edges, blending="additive", colormap="magma")
```

```python
config = MainConfig()
pprint(config)
```

```python
params_df = estimate_parameters_from_labels(labels, is_timelapse=True)
params_df["area"].plot(kind="hist", bins=100, title="Area histogram")
```

```python
config.segmentation_config.min_area = 3000
config.segmentation_config.max_area = 7000
config.segmentation_config.n_workers = 8

config.linking_config.max_distance = 40
config.linking_config.n_workers = 8

config.tracking_config.appear_weight = -1
config.tracking_config.disappear_weight = -1
config.tracking_config.division_weight = -0.1
config.tracking_config.power = 4
config.tracking_config.bias = -0.001
config.tracking_config.solution_gap = 0.0

pprint(config)
```

```python
track(
    detection=detection,
    edges=edges,
    config=config,
    overwrite=True,
)
```

```python
tracks_df, graph = to_tracks_layer(config)
labels = tracks_to_zarr(config, tracks_df)
```

```python
#viewer = napari.Viewer()
viewer.add_tracks(tracks_df[["track_id", "t", "y", "x"]].values, graph=graph)
#viewer.add_labels(labels)

#viewer.layers["Cellpose"].visible = False
#viewer.layers["edges"].visible = False
```

```python
output_dir = r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_converted_Z3\Tracking_trap1"

# Convert the output directory string to a Path object
output_dir_path = Path(output_dir)

# Export tracking results to CTC format
export.to_ctc(output_dir_path, config)
```

```python
xml_trackmate_output = export.to_trackmate(config)
```

```python
file_path = r'C:\Users\micha\Desktop\Master_Project\Trackmate\trackmate_output_olddata_celldiv.xml'

# Open the file in write mode and write the XML content
with open(file_path, 'w') as file:
    file.write(xml_trackmate_output)

print("XML file saved successfully at:", file_path)
```
