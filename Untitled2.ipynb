```python
import tifffile
import cv2
import numpy as np
import matplotlib.pyplot as plt

def chaccum(image, radius_range, method='phasecode', object_polarity='bright', edge_threshold=None):
    # Compute gradient using Sobel operator
    gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    gradient_img = np.sqrt(gradient_x**2 + gradient_y**2)
    
    # Normalize gradient_img to [0, 255]
    gradient_img_normalized = cv2.normalize(gradient_img, None, 0, 255, cv2.NORM_MINMAX)
    
    # Convert to 8-bit unsigned integer
    gradient_img_uint8 = np.uint8(gradient_img_normalized)
    
    # Get edge pixels
    if edge_threshold is None:
        edge_threshold, _ = cv2.threshold(gradient_img_uint8, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    edge_pixels = np.column_stack(np.where(gradient_img_uint8 > edge_threshold))

    # Identify different radii for votes
    if isinstance(radius_range, int):
        radius_range = [radius_range]
    elif len(radius_range) > 1:
        radius_range = np.arange(radius_range[0], radius_range[1] + 0.5, 0.5)

    if object_polarity == 'dark':
        radius_range = -radius_range

    # Compute weights for votes
    if method == 'twostage':
        weights = 1 / (2 * np.pi * np.abs(radius_range))
    elif method == 'phasecode':
        if len(radius_range) > 1:
            phi = ((np.log(radius_range) - np.log(radius_range[0])) / (np.log(radius_range[-1]) - np.log(radius_range[0])) * 2 * np.pi) - np.pi
        else:
            phi = 0
        opca = np.exp(1j * phi)
        weights = opca / (2 * np.pi * np.abs(radius_range))
    else:
        raise ValueError("Invalid method. Choose 'twostage' or 'phasecode'.")

    # Computing the accumulator array
    accum_matrix = np.zeros_like(image, dtype=np.complex128)

    for radius in radius_range:
        xc = edge_pixels[:, 1] - radius * gradient_x[edge_pixels[:, 0], edge_pixels[:, 1]] / gradient_img[edge_pixels[:, 0], edge_pixels[:, 1]]
        yc = edge_pixels[:, 0] - radius * gradient_y[edge_pixels[:, 0], edge_pixels[:, 1]] / gradient_img[edge_pixels[:, 0], edge_pixels[:, 1]]

        xc_rounded = np.round(xc).astype(int)
        yc_rounded = np.round(yc).astype(int)

        valid_indices = np.where((xc_rounded >= 0) & (xc_rounded < image.shape[1]) & (yc_rounded >= 0) & (yc_rounded < image.shape[0]))

        xc_rounded = xc_rounded[valid_indices]
        yc_rounded = yc_rounded[valid_indices]
        weights_valid = weights[valid_indices]

        accum_matrix[yc_rounded, xc_rounded] += weights_valid

    return accum_matrix, gradient_img


# Example usage
tifffile_image = tifffile.imread(r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_converted_Z1\Cropped_traps_directory\trap_1_all_frames_series.tiff")
image = tifffile_image[1, 0, :, :]

radius_range = [30, 120]  # Set your desired radius range
accum_matrix, gradient_img = chaccum(image, radius_range)

# Normalize the accumulator matrix for visualization
accum_matrix_normalized = np.abs(accum_matrix) / np.max(np.abs(accum_matrix))

# Display the images using Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(accum_matrix_normalized, cmap='gray')
plt.title('Normalized Accumulator Matrix')

plt.subplot(1, 2, 2)
plt.imshow(gradient_img, cmap='gray')
plt.title('Gradient Image')

plt.show()
```
