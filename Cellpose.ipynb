```python
import numpy as np
from cellpose import models, io
from skimage import io as skio, measure
from os.path import join, basename
import os

# Directory containing the input images
input_directory = r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Individual_cropped_traps_directory\Cropped_traps_position_5"

# List all files in the input directory
files = [f for f in os.listdir(input_directory) if f.endswith('.tiff')]

# DEFINE CELLPOSE MODEL
# model_type='cyto' or model_type='nuclei'
model = models.Cellpose(gpu=False, model_type='cyto2')

# or if you have different types of channels in each image
channels = [[0, 0]]

# Set custom parameters
custom_params = {
    'diameter': 60.0,
    'interp': 'lanczos3',
    'cellprob_threshold': 0.1,
    'flow_threshold': 0.4,
    'min_size': 30
}

# Output directory for unique masks
output_dir = r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Unique_masks_directory"
os.makedirs(output_dir, exist_ok=True)

#Add tqdm for the progress bar
for n, filename in tqdm(enumerate(files, start=1), total=len(files), desc="Processing Images"):
    # Read the image
    img = io.imread(join(input_directory, filename))

    # Perform cell segmentation
    masks, _, _, _ = model.eval(img, channels=channels, **custom_params)

    # Save unique cell labels as numpy array TIFF file
    unique_labels = measure.label(masks)

    # Save the unique mask with a specific filename
    output_filename = f"unique_mask_position_5_frame_{n}.tiff"
    output_filepath = join(output_dir, output_filename)
    skio.imsave(output_filepath, unique_labels.astype(np.uint16), plugin='tifffile')
```

```python
import napari
from skimage import io as skio
from skimage import measure, segmentation
import numpy as np

original_image_path = r"C:\Users\micha\Desktop\Master_Project\original_frame_392_cropped\trap_2.tiff"
unique_labels_path = r"C:\Users\micha\Desktop\Master_Project\original_frame_392_cropped\unique_cell_labels.tiff"

# Read the images
original_image = skio.imread(original_image_path)
unique_labels = skio.imread(unique_labels_path)

# Set a threshold for minimum area
min_area_threshold = 100

# Compute region properties
props = measure.regionprops(unique_labels)

# Get centroids and labels of labeled regions that meet the area threshold
filtered_props = [(prop.label, prop.centroid) for prop in props if prop.area >= min_area_threshold]

# Remove labels touching the border
filtered_labels_no_border = segmentation.clear_border(unique_labels)

# Get centroids and labels of remaining labeled regions
filtered_props_no_border = [(label, prop.centroid) for label, prop in enumerate(measure.regionprops(filtered_labels_no_border), start=1)]

# Open Napari viewer
viewer = napari.Viewer()

# Add layers to the viewer
viewer.add_image(original_image, name='Original Image')

# Add unique cell labels with a colormap
viewer.add_labels(unique_labels, name='Unique Cell Labels')

# Add red crosses at the centroids of filtered labels
viewer.add_points(
    np.array([centroid for label, centroid in filtered_props]),
    size=20,
    face_color='white',
    edge_color='white',
    symbol='o',
    name='Filtered Centroids (Area Threshold)'
)

# Add green crosses at the centroids of remaining labels after removing border touching
viewer.add_points(
    np.array([centroid for label, centroid in filtered_props_no_border]),
    size=20,
    face_color='green',
    edge_color='green',
    symbol='o',
    name='Filtered Centroids (Area + Border Removal)'
)

napari.run()
```

```python
import napari
from skimage import io as skio
from skimage import measure, segmentation
import numpy as np

original_image_path = r"C:\Users\micha\Desktop\Master_Project\original_frame_392_cropped\trap_2.tiff"
unique_labels_path = r"C:\Users\micha\Desktop\Master_Project\original_frame_392_cropped\unique_cell_labels.tiff"

# Read the images
original_image = skio.imread(original_image_path)
unique_labels = skio.imread(unique_labels_path)

# Set a threshold for minimum area
min_area_threshold = 100

# Remove labels based on area threshold
filtered_labels_no_border = segmentation.clear_border(unique_labels)
props = measure.regionprops(filtered_labels_no_border)

# Remove labels that don't meet the minimum area threshold
filtered_labels = np.zeros_like(unique_labels)
for prop in props:
    if prop.area >= min_area_threshold:
        filtered_labels[prop.coords[:, 0], prop.coords[:, 1]] = prop.label

# Open Napari viewer
viewer = napari.Viewer()

# Add layers to the viewer
viewer.add_image(original_image, name='Original Image')

# Add unique cell labels with a colormap

# Add filtered unique cell labels with a colormap
viewer.add_labels(filtered_labels, name='Filtered Unique Cell Labels')

napari.run()

props = measure.regionprops(filtered_labels)

# Iterate over each labeled region (cell) and access its properties
for prop in props:
    label = prop.label
    area = prop.area
    centroid = prop.centroid
    bbox = prop.bbox
    eccentricity = prop.eccentricity 

    # Check if the area meets the threshold
    if area >= min_area_threshold:
        # Print or use the properties as needed for your analysis
        print(f"Label: {label}, Area: {area}, Centroid: {centroid}, Bounding Box: {bbox}, Eccentricity: {eccentricity}")
```

```python
import napari
from skimage import io as skio
from skimage import measure, segmentation
import numpy as np
import os

# Directory containing the original images
original_images_directory = r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Individual_cropped_traps_directory\Cropped_traps_position_5"

# Directory containing the unique cell masks
masks_directory = r"C:\Users\micha\Desktop\MM_data_Pipeline_results_all\MM_pipeline_results_tiff_file_converted_2\Unique_masks_directory"

# Set thresholds for minimum and maximum area, and maximum eccentricity
min_area_threshold = 100
max_area_threshold = 10000  # Adjust as needed
max_eccentricity_threshold = 0.8  # Adjust as needed

# Read all original images and masks
original_images = []
filtered_masks = []

# List all files in the directory and sort them
files = sorted(os.listdir(original_images_directory))

for filename in files:
    if filename.endswith('.tiff'):
        original_image_path = os.path.join(original_images_directory, filename)
        unique_labels_path = os.path.join(masks_directory, f"unique_cell_labels_{filename}")

        # Read the images
        original_image = skio.imread(original_image_path)
        unique_labels = skio.imread(unique_labels_path)

        # Remove labels based on area threshold
        filtered_labels_no_border = segmentation.clear_border(unique_labels)
        props = measure.regionprops(filtered_labels_no_border)

        # Remove labels that don't meet the area and eccentricity thresholds
        filtered_labels = np.zeros_like(unique_labels)
        for prop in props:
            if min_area_threshold <= prop.area <= max_area_threshold and prop.eccentricity <= max_eccentricity_threshold:
                filtered_labels[prop.coords[:, 0], prop.coords[:, 1]] = prop.label

        original_images.append(original_image)
        filtered_masks.append(filtered_labels)

# Open Napari viewer
viewer = napari.Viewer()

# Add original images in a horizontal stack
viewer.add_image(np.hstack(original_images), name='Original Images', blending='additive')

# Add filtered unique cell masks in a separate layer
viewer.add_labels(np.hstack(filtered_masks), name='Filtered Unique Cell Masks')

napari.run()
```
